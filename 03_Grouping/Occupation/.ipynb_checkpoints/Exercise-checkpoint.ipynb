### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|', index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# first confirm all possible gender values
users['gender'].value_counts()
```

```python
occs = users.groupby('occupation').sum()
occs['M'] = 0
occs['F'] = 0
for occupation in list(users['occupation'].unique()):
    for gender in users[users['occupation']==occupation]['gender']:
        if gender == 'M':
            occs.loc[occupation,'M'] += 1
        else:
            occs.loc[occupation,'F'] += 1
occs['m_to_f_ratio'] = occs['M']/occs['F']
occs['percent_male'] = (occs['M']/(occs['M']+occs['F']))*100
occs.sort_values(by='percent_male',ascending=False,inplace=True)
occs
```
